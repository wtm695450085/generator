```python
import numpy as np
import pandas as pd

np.random.seed(0)  

AA1 = np.random.uniform(8.1,15.7, size=600) # losowe fload
AA2 = np.random.randint(12,25, size=600)    # losowe int

# --------------- generator 3 składnikowych znaków losowych powtarzających się---
import string
import random

def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))

STR_1 = []
for a in range(600):
    STR_1.append(id_generator(3, "3etrY"))
# ----------------------------------------------------------------
STR_2 = []
for a in range(600):
    STR_2.append(id_generator(2, "AKLS3"))
# -----------------------------------------------------------------------
AA3 = np.random.uniform(0.1,2.7, size=600) # losowe fload
AA4 = np.random.randint(1222,2235, size=600)    # losowe int
# ----------------------------------------------------------------
STR_3 = []
for a in range(600):
    STR_3.append(id_generator(4, "KPL"))
# -----------------generator wartości z WARTOŚCIAMI ODSTAJĄCYMI---------------

def generate(median=630, err=12, outlier_err=100, size=80, outlier_size=10):
    errs = err * np.random.rand(size) * np.random.choice((-1, 1), size)
    data = median + errs

    lower_errs = outlier_err * np.random.rand(outlier_size)
    lower_outliers = median - err - lower_errs

    upper_errs = outlier_err * np.random.rand(outlier_size)
    upper_outliers = median + err + upper_errs

    data = np.concatenate((data, lower_outliers, upper_outliers))
    np.random.shuffle(data)

    return data

# --------PIERWSZY SZEREG Z OUTLAYERAMI -----
# wielkość próby 580+(2*10)=600
AA5 = generate(median = 630, err = 12, outlier_err = 100, size = 580, outlier_size = 10)

# --------DRUGI SZEREG Z OUTLAYERAMI -----
# wielkość próby 590+(2*5)=600
AA6 = generate(median = 30, err = 2, outlier_err = 10, size = 590, outlier_size = 5) 

## ----------- wartość wynikowa -------------------------------------

AA7 = np.random.uniform(88.1,115.7, size=600) # losowe fload
```

```python
df = pd.DataFrame({
    'AA1': AA1,
    'AA2': AA2,
    'AA3': AA3,
    'AA4': AA4,
    'STR_1': STR_1,
    'STR_2': STR_2,
    'STR_3': STR_3,
    'AA5': AA5,
    'AA6': AA6,
    'AA7': AA7,
})
df
```
